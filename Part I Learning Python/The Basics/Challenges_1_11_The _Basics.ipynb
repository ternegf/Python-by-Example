```python
# *****
# 001 *
#************************************************************
# Ask for the user’s first name and display the output message
# Hello [First Name].
# ************************************************************
def alpha(string):
    """
    Returns just the alpha characters in the string that was
    passed (i.e. a-Z and A-Z).
    
    Arg(s):
        String
        
    Returns:
        String that just has the alpha character of the string
        that was passed as an argument.
        
        For example:
        
        alpha('@28Al9)i [c]]e')
        
        returns Alice
    """
    alpha_string = ''
    for char in string:
        if char.upper() >= 'A' and char.upper() <= 'Z':
            alpha_string = alpha_string + char
            
    return alpha_string
            
name = input ('Enter your name : ')
name = alpha(name)
name = name.title()

print(f'Your name is {name}.')
```

```python
# *****
# 002 *
#************************************************************
# Ask for the user’s first name and then ask for their surname
# and display the output message:
#
# Hello [First Name] [Surname]
# ************************************************************
def process_name(name):
    """
    Processes a name string, using the alpha function defined
    in challenge #1 to remove any non-alpha characters.
    
    Arg(s):
        String to be proccessed.
        
    Returns:
        Processed string. 
    """
    name = alpha(name)
    name = name.title()
    return name

first = input('First Name : ')
first = process_name(first)

last  = input('Last Name  :')
last  = process_name(last)

print(f'\nHello {first} {last}!')

```

```python
# *****
# 004 *
# ************************************************************
# Ask the user to enter two numbers, add them together and
# displad the answer as:
#
# The total is [answer]
# ************************************************************
 
def get_valid_number(prompt):
    valid_number = False
    while not valid_number:
        try:
            number_string = input(prompt)
            number_float  = float(number_string)
            valid_number = True
        except ValueError:
            valid_number = False

    return number_float

# *****************************************************+
# Convert the strings entered to floating point numbers.
# ******************************************************

number_one = get_valid_number('Enter a number.      : ')    
number_two = get_valid_number('Enter another number : ')

answer     = number_one * number_two

decimal_part = answer % 1
#print(f'{decimal_part=}')
if decimal_part == 0.0:
    # ***************************************************
    # If the answer has no decimal places convert it from
    # floating point to an integer.
    # ***************************************************
    answer = int(answer)
    
print(f'\nThe total is {answer}' )
```

```python
# *****
# 005 *
# ***********************************************************
# Ask the user to enter 3 numbers. Add the first two together
# then multiply the result by the third.
# ***********************************************************
def find_operator(operator):
    operator_list = ['(',
                '+',
                '-',
                '*',
                '/',
                '**',
                ')']
    
    found = False
    for op in operator_list:
        
        #print(f'if {operator} == {op}')
        if operator == op:

            found = True
            break
        
    return found        
    
stack = []
#special_characters = '(+-*/)'

string_list = []

more_characters=True
while more_characters:
    value = input('Enter value : ')
    if value == '':
        more_characters=False
    else:
        string_list.append(value)
        
#print(string_list)

for oper in string_list:
    #print (oper)    
    if not find_operator(oper):
        operand = oper
        stack.append(operand)
        #print(f'{operand=}')
    else:
        operator = oper
        #print(f'{operator=}')

        operand_1 = float(stack.pop(0))
        operand_2 = float(stack.pop(0))
        if operator == '+':
            result = operand_1 + operand_2
            #print(f'{result} = {operand_1} + {operand_2}')
            stack.append(str(result))
        elif operator == '-':
            result = operand_1 - operand_2
            #print(f'{result} = {operand_1} - {operand_2}')
            stack.append(str(result))
        elif operator == '*':
            result = operand_1 * operand_2
            #print(f'{result} = {operand_1} * {operand_2}')
            stack.append(str(result))
        elif operator == '**':
            result = operand_1 ** operand_2
            #print(f'{result} = {operand_1} ** {operand_2}')
            stack.append(str(result))
        elif operator == '/':
            result = operand_1 / operand_2
            #print (result)
            #print(f'{result} = {operand_1} / {operand_2}')
            stack.append(str(result))        
    
answer = float(stack.pop(0))

decimal_part = answer % 1
if decimal_part == 0.0:
    # ***************************************************
    # If the answer has no decimal places convert it from
    # floating point to an integer.
    # ***************************************************
    answer = int(answer)

print(f'The answer is {answer}')
```

```python
# *****
# 010  *
#***************************************************
# There are 2.204 pounds in a kilogram, ask the user
# to  enter  a weight in kilograms and convert it to
# pounds.
# **************************************************
# Note: This solution could easily be made generic &
#       used to convert other things, such as meters
#       to feet and inches.
# **************************************************

# *********************************************************************
# F U N C T I O N S
# *********************************************************************

# –—------------------------------------------------------------------
# convert
# –—------------------------------------------------------------------
def convert(weight):
    '''Convert weight in kilograms to weight in stones, pounds and
    ounces.
    
    Arguement(s)
        Weight in kg
        
    Returns
        Dictionary with values for the keys  "ST", "LB" and "OZ".
  '''
    # ***************************************
    # Set the conversion values as constants.
    # **************************************
    KG2ST = 0.1574730444
    ST2LB = 14.0
    LB2OZ = 16.0
    
    CONVERT = (KG2ST, ST2LB, LB2OZ)
    
    UNITS    = ('ST','LB','OZ')
    float: conversion
    converts = {}
    
    # ********************************************************
    # Iterate  through the conversions, storing the converted
    # values  in the dictionary, converts, with a key of what
    # we  are  converting  to.  Repeat for the remainder with
    # the next conversion unit.
    # ********************************************************
    for i, conversion in enumerate(CONVERT):
        unit = UNITS[i]
        weight    = weight * conversion
        converts[unit] = int(weight)
        weight = weight - int(weight)
        
    # --------------        
    return(converts)
    # --------------
    
# *********************************************************************
# M A I N  P R O G R A M
# *********************************************************************

# **********************************************************
# The descriptions we're going to use for each of the units.
# **********************************************************
unit_descr = {'ST':'stones','LB':'pounds','OZ':'ounces'}

# ***********
# Get weight.
# ***********
weight_kg = float(input('Weight in kg : '))

# ******************
# 2. Convert weight.
# ******************
conversions = convert(weight_kg)

# ****************************************
# 3. Display the weight entered in stones,
#    pounds & ounces.
# ****************************************
delimiter = ''
out_line = f'\n{weight_kg} kg is equal to'

for i, key in enumerate(conversions.keys()):
    out_line = out_line + f'{delimiter} {conversions[key]} {unit_descr[key]}'
    if i+1 < len(conversions) - 1:
        delimiter = ','
    else:
        delimiter = ' and'
    
out_line = out_line + '.'
print(f'{out_line}\n')

```
