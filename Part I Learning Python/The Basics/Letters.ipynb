```python
import matplotlib.pyplot as plt
import pandas as pd

string = input('Enter string : ')

#string = 'AZ'

letter_counts = {}
for character in string.upper():
    index = ord(character)
    if index >=65 and index <= 90:
        if character in letter_counts:
            letter_counts[character] += 1
        else:
            letter_counts[character] = 1

# Convert the dictionary to a Pandas DataFrame
letter_counts_df = pd.DataFrame.from_dict(letter_counts, orient='index', columns=['Count'])

# Sort the DataFrame by letter
letter_counts_df = letter_counts_df.sort_index()

# Create a bar chart to visualize the letter distribution
plt.bar(letter_counts_df.index, letter_counts_df['Count'])
plt.xlabel('Letter')
plt.ylabel('Count')
plt.title(string)
#plt.title('Hey Jude')

# Show the chart

plt.show()
```
