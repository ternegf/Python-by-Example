```python
# *********************************************************************
#                                                                     *
# Program      : Challenge_005_v1_01.py                               *
# Written by   : George Ternent                                       *
# Date.        : 13-Nov-2023.                                         *
#                                                                     *
# Description : Ask  the  user  to enter 3 numbers. Add the first two *
#               together then multiply the result by the third.       *
#                                                                     *
# *********************************************************************
# Function(s)                                                         *
# *********************************************************************
#                                                                     *
# Name             Description                                        *
# ****             ***********                                        *
# find_operator   Search  for an operator in a string, returning True *
#                 if found and False if not.                          *
#                                                                     *
# *********************************************************************
# History                                                             *
# *********************************************************************
#                                                                     *
# Ref   Changed by        Date          Change(s)                     *
# ****  ***********       ****         **********                     *
# 0.01  George Ternent.   13-Nov-2023. Initial Version.               *
#                                                                     *
# *********************************************************************

# *********************************************************************
# F U N C T I O N S
# *********************************************************************

# –—------------------------------------------------------------------
# find_operator
# –—------------------------------------------------------------------
def find_operator(operator):
    operator_list = ['(',
                '+',
                '-',
                '*',
                '/',
                '**',
                ')']
    
    found = False
    for op in operator_list:
        
        #print(f'if {operator} == {op}')
        if operator == op:

            found = True
            break
        
    # ----------
    return found
    # ----------

# *********************************************************************
# M A I N  P R O G R A M
# *********************************************************************

```
