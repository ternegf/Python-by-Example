```python
def calculate_bmi(weight_kg, height_m):
    '''
    Calculate BMI
    
    Argument(s):
        weight in kilograms
        height in metres
        
    Returns:
        Body Mass Index
    '''
    bmi = weight_kg / (height_m ** 2)

    return bmi


def compose_message(bmi, sex):
    ''' 
    Compose BMI message.
    
    Argument(s):
        Body Mass Index
        
    Returns:
        Message.
    '''
    
    message_cat_m = { "UNDERWEIGHT" : " is underweight. Slim Jim!",
                      "HEALTHY"     : " is a healthy weight. Cool!",
                      "OVERWEIGHT"  : " is overweight. Don't eat all the pies!",
                      "OBESE"       : " is obese. Bunter!" }
    
    message_cat_f = { "UNDERWEIGHT" : " is underweight. Twiggy!",
                      "HEALTHY"     : " is a healthy weight. Cool!",
                      "OVERWEIGHT"  : " is overweight. Lay off the chocolates!",
                      "OBESE"       : " is obese. Oi, Hattie Jaques!" }
    if sex == 'M':
        message_cat = message_cat_m
    else:
        message_cat = message_cat_f
        
    if bmi < 18.5:
        return name + message_cat["UNDERWEIGHT"]
    elif bmi <= 25.0:
        return name + message_cat["HEALTHY"]
    elif bmi <= 29.9:
        return name + message_cat["OVERWEIGHT"]
    else:
        return name + message_cat["OBESE"]

    
def bmi_calculator (name, weight_kg, height_m, sex):
    '''
    Calculate and display BMI, with a suitable message.
     
    Arguments:
        name
        weight in kilograms
        height in metres
        
    Returns:
        Message to display.
    '''
    bmi = calculate_bmi(weight_kg, height_m)
    
    messages = [name]
    messages.append(f"Weight: {weight} kg, Height: {height_m} m, bmi: {round(bmi,1)} ")
    messages.append(compose_message(bmi, sex) + "\n")
    return messages


names = ["Harry Cool", "Harry Cool's sister", "Harry Cool's brother"]
sexes = ["M","F","M"]
weights = []
heights = []

df_data = {"NAME":["Stig Lundquist", "Quentin Gelder", "Alice","Linda","Bunman"],
           "SEX":["M","M","F","F","M"],
           "WEIGHT":[79.0, 79.0, 74.0, 81.8, 79.83],
           "HEIGHT":[1.91, 1.91, 1.78, 1.57, 1.8288]}


```
